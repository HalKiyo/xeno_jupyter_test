```python
def l_coordinate_to_tuple(lcoordinate):
    b = 4320
    lat_l = ((lcoordinate - 1) // b)
    lon_l = (lcoordinate) % b - 1
    return (lat_l, lon_l)
```

```python
result = np.empty(upp.shape, dtype=tuple)
vfunc = np.vectorize(l_coordinate_to_tuple, otypes=[tuple])
mask = ~np.isnan(upp)
result[mask] = vfunc(upp[mask])
print(result.shape)
```

```python
#upp = lonlat[900:930, 3355:3375]
# diff = 900, 3355
# start = 910, 3365

def loop(lst):
    fix_xxx = 3355
    fix_yyy = 900
    latest = lst[-1]
    print(latest)
    start_y = latest[0]
    start_x = latest[1]
    next_dst = result[start_y, start_x]
    x = int(next_dst[1] - fix_xxx)
    y = int(next_dst[0] - fix_yyy)
    lst.append([y, x])
    return lst
    
```

```python
lst = [[5, 5]]
for _ in range(10):
    lst = loop(lst)
print(lst)
```
