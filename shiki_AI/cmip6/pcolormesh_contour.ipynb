```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.05)
y = np.arange(0, 10, 0.05)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

plt.pcolormesh(X, Y, Z, cmap='hsv')

pp = plt.colorbar(orientation="vertical")
pp.set_label("Label", fontsize=24)

plt.xlabel('X', fontsize=24)
plt.ylabel('Y', fontsize=24)

plt.show()
```

```python
plt.imshow(Z, cmap='hsv', origin='lower')
plt.colorbar(orientation="vertical")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
cont = plt.contour(X, Y, Z, 5, colors=['black'])
cont.clabel(fmt='%1.1f', fontsize=14)

pp = plt.colorbar(orientation="vertical")
pp.set_label("label", fontsize=24)

plt.xlabel('X', fontsize=24)
plt.ylabel('Y', fontsize=24)

plt.show()
```

```python
cont = plt.contour(X, Y, Z, 5, colors=['black'])
cont.clabel(fmt='%1.1f', fontsize=14)

plt.xlabel('X', fontsize=24)
plt.ylabel('Y', fontsize=24)

plt.imshow(Z, cmap='cool', origin='lower')
pp = plt.colorbar(orientation="vertical")
pp.set_label("label", fontsize=24)

plt.show()
```
