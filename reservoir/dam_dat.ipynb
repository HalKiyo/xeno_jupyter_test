```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#貯水池の需要がdam_dem_に記録される．
month = 100 #January
dirname = '/home/kajiyama/H08/H08_20230612/dam/dat'
path = dirname + f"/dam_dem_/WFDEN_C_0000{month:04}.hlf"
dtype = 'float32' # 4バイト
data = np.fromfile(path, dtype=dtype)
lonlat = data.reshape(360, 720)
```

```python
plt.imshow(lonlat, vmin=0, vmax=1)
plt.colorbar()
```

```python
tha = lonlat[134:167, 550:580]
condition = np.logical_and(0 < tha, tha < 1e20)
indices = np.where(condition)
print(indices[0])
print(indices[1])
```

```python
plt.imshow(tha, vmin=0, vmax=30000)
plt.colorbar()
```

```python
for i in range(len(indices[0])):
    n = i
    value = tha[indices[0][n], indices[1][n]]
    print(value)
```
