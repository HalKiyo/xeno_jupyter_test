```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
%matplotlib inline
```

```python
fig = plt.figure()
projection = ccrs.PlateCarree(central_longitude=180)
ax = fig.add_subplot(1,1,1, projection = projection)

ax.add_feature(cfeature.LAND, 
               edgecolor= 'face',
               facecolor='dimgrey')
ax.add_feature(cfeature.OCEAN, 
               edgecolor='face',
               facecolor='darkgrey')
```

```python
path = '/work/kajiyama/cdo/cmip6/AWI-ESM-1-1-LR/snc/snc_1850-1850.nc'
from os.path import exists
exists(path)
```

```python
from netCDF4 import Dataset
ds = Dataset(path, 'r')
snc = ds.variables['snc'][:]
```

```python
import colorcet as cc
plt.register_cmap('cc_blues_r', cc.cm.blues_r)
```

```python
fig = plt.figure()
projection = ccrs.PlateCarree(central_longitude=180)
ax = fig.add_subplot(1,1,1, projection = projection)

ax.add_feature(cfeature.LAND, 
               edgecolor= 'face',
               facecolor='dimgrey')
ax.add_feature(cfeature.OCEAN, 
               edgecolor='face',
               facecolor='darkgrey')

img_extent = (-180, 180, -90, 90)
cmap = plt.get_cmap('cc_blues_r', 256)
mat = ax.matshow(snc[0,::-1, :], 
                 origin='upper',
                 extent=img_extent,
                 transform = projection,
                 cmap=cmap)
cbar = fig.colorbar(mat, ax=ax, orientation='horizontal')
```

```python
import numpy as np
import matplotlib.colors as colors

fig = plt.figure()
projection = ccrs.PlateCarree(central_longitude=180)
ax = fig.add_subplot(1,1,1, projection = projection)

ax.add_feature(cfeature.LAND, 
               edgecolor= 'face',
               facecolor='dimgrey')
ax.add_feature(cfeature.OCEAN, 
               edgecolor='face',
               facecolor='darkgrey')

img_extent = (-180, 180, -90, 90)
cmap = plt.get_cmap('cc_blues_r', 256)
cmap_data = cmap(np.arange(cmap.N))
cmap_data[0, 3] = 0
my_cmap = colors.ListedColormap(cmap_data)
mat = ax.matshow(snc[0,::-1, :], 
                 origin='upper',
                 extent=img_extent,
                 transform = projection,
                 cmap=my_cmap)
cbar = fig.colorbar(mat, ax=ax, orientation='horizontal')
```
