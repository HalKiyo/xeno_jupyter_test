```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import pprint
h08dir = '/home/kajiyama/H08/H08_20230612'
SUF = '.gl5'
dtype = 'float32'
gl5shape = (2160, 4320)
hlfshape = (360, 720)
```

```python
# canal file
origin = np.fromfile('/home/kajiyama/H08/H08_20230612/map/org/Saritha/4_existing_origin.hlf', dtype='float32').reshape(hlfshape)
origin_masked = np.ma.masked_where(origin == 0, origin)

projection = ccrs.PlateCarree(central_longitude=0)
img_extent = (-180, 180, -90, 90)
fig = plt.figure(figsize=(50,50))
ax = plt.subplot(projection=projection)
ax.coastlines()
im = ax.imshow(origin_masked, extent=img_extent, transform=projection, cmap='rainbow')
plt.colorbar(im)
```

```python
# tokyo region
# "138 141 34 38"
one = 25 
two = 40
three = 25
four = 35
upperindex = (90-four)*2
lowerindex = (90-three)*2
leftindex = (180+one)*2
rightindex = (180+two)*2
print(upperindex, lowerindex, leftindex, rightindex)

# crop
origin_cropped = origin_masked[int(upperindex):int(lowerindex), int(leftindex):int(rightindex)]

# masked
print(np.max(origin_cropped))

projection = ccrs.PlateCarree(central_longitude=0)
img_extent = (one, two, three, four)
fig = plt.figure()
ax = plt.subplot(projection=projection)
ax.coastlines()
im = ax.imshow(origin_cropped, extent=img_extent, transform=projection, cmap='rainbow')
plt.colorbar(im)
```
