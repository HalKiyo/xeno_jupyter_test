```python
def worldmap(img):
    projection = ccrs.PlateCarree(central_longitude=0)
    img_extent = (-180, 180, -90, 90)

    fig = plt.figure()
    ax = plt.subplot(projection=projection)
    ax.coastlines()
    im = ax.imshow(img,
                   origin='upper',
                   extent=img_extent,
                   transform=projection,
                   cmap='WhiteBlueGreenYellowRed',
                   vmin=0, vmax=3e-4)
    fig.colorbar(im, ax=ax)
```

```python
def monthly(year = 2011):
    raindir = '/home/kajiyama/H08/H08_20230612/met/dat/Rainf___'
    maskpath = '/home/kajiyama/H08/H08_20230612/map/dat/lnd_msk_/lndmsk.WFDEI.hlf'
    dtype = 'float32' # 4バイト
    box = []
    for i in range(1, 13):
        rainpath = raindir + f"/W5E5____{year}{i:02}00.hlf"
        data = np.fromfile(rainpath, dtype=dtype)
        lonlat = data.reshape(360, 720)
        total = np.sum(lonlat)
        box.append(total)
    return box
```

```python
box = monthly(year=2011)
print(box)
```

```python
sam79 = monthly(year=1979)
sam89 = monthly(year=1989)
sam99 = monthly(year=1999)
sam09 = monthly(year=2009)
sam11 = monthly(year=2011)
sam19 = monthly(year=2019)

x = np.arange(12)
plt.plot(x, sam79, label='1979')
plt.plot(x, sam89, label='1989')
plt.plot(x, sam99, label='1999')
plt.plot(x, sam09, label='2009')
plt.plot(x, sam19, label='2019')
plt.legend()
```
